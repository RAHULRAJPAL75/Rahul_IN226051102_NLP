**Task 1: Conceptual Understanding**

**1.What is the difference between "Love" and "love" in NLP?**

In Natural Language Processing, “Love” and “love” are treated as different tokens because of case sensitivity. This means the model sees uppercase and lowercase words as separate. For example, “Love” might appear at the beginning of a sentence or represent a proper noun, while “love” is generally used as a regular word. Because of this difference, models may interpret them differently. To avoid such inconsistencies, text is often converted to lowercase during preprocessing.

**2.What happens if stopwords are not removed?**

If stopwords are not removed in NLP, the dataset will include many frequently used words like “the”, “is”, and “and” that do not contribute much to the actual meaning. This can add unnecessary noise and make it harder for the model to identify important information. It may also slow down processing and reduce overall efficiency, which can impact the model’s performance.

**3.Provide two real-world scenarios where removing stopwords can be harmful.**

Removing stopwords is not always beneficial. In voice-based systems, small words can be crucial for meaning—for example, removing “off” from “turn off the light” could lead to the wrong action. Another case is text summarization, where stopwords help maintain sentence structure and readability. Eliminating them can make the output feel unnatural or difficult to understand.

**4.Explain the difference between stemming and lemmatization.**

Stemming and lemmatization are techniques used to reduce words to their base form in NLP, but they differ in approach. Stemming works by trimming word endings, which is fast but can produce incorrect forms (e.g., “running” → “runn”). Lemmatization, however, considers the word’s meaning and context to return the correct base form (e.g., “went” → “go”). In short, stemming is quicker but less precise, while lemmatization is more accurate but computationally heavier.

**Task 3: Stress Testing**

```python
import re
def preprocess_text(text):
    text = text.casefold()
    combined_pattern = r'(http\S+|www\S+|\S+@\S+|\b\d{7,}\b)'
    text = re.sub(combined_pattern, ' ', text)
    text = ''.join([ch if not ch.isdigit() else ' ' for ch in text])
    def fix_repetition(word):
        return re.sub(r'(.)\1{2,}', r'\1\1', word)

    text = " ".join(fix_repetition(word) for word in text.split())
    cleaned_chars = []
    for ch in text:
        if ch.isalpha() or ch.isspace():
            cleaned_chars.append(ch)
        else:
            cleaned_chars.append(' ')
    text = "".join(cleaned_chars)
    text = " ".join(text.split())
    final_tokens = []
    for word in text.split():
        if len(word) > 2:
            final_tokens.append(word)
        elif word in ("no", "not", "ok"):
            final_tokens.append(word)
    cleaned_text = " ".join(final_tokens)

    return final_tokens, cleaned_text
sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]
print("\n--- NLP Preprocessing Stress Test Results ---\n")
for i, text in enumerate(sentences, 1):
    tokens, cleaned = preprocess_text(text)
    print(f"Case {i}:")
    print("Input:   ", text)
    print("Tokens:  ", tokens)
    print("Output:  ", cleaned)
    print("-" * 50)
```

**Task 4: Token Analytics**

```python
samples = [
    "Hurry!!! Grab 50% discount today only!!!",
    "Soooo happy with this amazing service 😊😊",
    "Terrible experience... will not recommend!!!",
    "Contact us at support@company.com for help",
    "This course is really really helpful!!!",
    "Check this out: www.example.org now!",
    "Noooooo way this actually worked!!!",
    "Ok ok I understand now",
    "Earn $$$ working from home!!!",
    "I am not satisfied with the results"
]
print("\n--- Token Insights Report ---\n")
analysis_data = []
for idx, sentence in enumerate(samples, 1):
    tokens, cleaned_text = preprocess_text(sentence)
    token_count = len(tokens)
    unique_count = len(set(tokens))
    richness = round(unique_count / token_count, 2) if token_count else 0
    char_density = round(sum(len(t) for t in tokens) / (len(cleaned_text) + 1), 2)
    analysis_data.append({
        "id": idx,
        "text": sentence,
        "tokens": tokens,
        "count": token_count,
        "unique": unique_count,
        "richness": richness,
        "density": char_density
    })

    print(f"Case {idx}:")
    print("Input Text:      ", sentence)
    print("Processed Tokens:", tokens)
    print("Token Count:     ", token_count)
    print("Unique Words:    ", unique_count)
    print("Lexical Score:   ", richness)
    print("Char Density:    ", char_density)
    print("-" * 50)
most_noisy = sorted(analysis_data, key=lambda x: (x["count"], x["richness"]))[0]
most_informative = sorted(analysis_data, key=lambda x: (x["richness"], x["count"]), reverse=True)[0]
print("\n--- Final Insights ---\n")
print("Most Noisy Sentence:")
print(f"Case {most_noisy['id']}: {most_noisy['text']}")
print("\nMost Informative Sentence:")
print(f"Case {most_informative['id']}: {most_informative['text']}")
```

**Task 5: Frequency Analysis**

```python
from collections import defaultdict
import re
def clean_and_tokenize(text):
    text = text.casefold()
    text = re.sub(r'(http\S+|www\S+|\S+@\S+|\b\d{6,}\b)', ' ', text)
    filtered = []
    for ch in text:
        if ch.isalpha() or ch.isspace():
            filtered.append(ch)
        else:
            filtered.append(' ')
    text = "".join(filtered)
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)
    text = " ".join(text.split())
    tokens = []
    for word in text.split():
        if len(word) > 2 or word in ("no", "not", "yes"):
            tokens.append(word)

    return tokens
data = [
    "Limited time offer!!! Get free coupons now!!!",
    "I am sooo happy with this purchase 😍",
    "Bad experience... not worth the money",
    "Reach us at helpdesk@service.com anytime",
    "This product quality is really really good",
    "Click here: www.fakeoffer.com and win prizes",
    "Nooooo I don't like this at all",
    "Ok yes I got the update",
    "Earn money fast $$$ from home!!!",
    "I am not satisfied with customer support"
]

freq_dict = defaultdict(int)

for sentence in data:
    tokens = clean_and_tokenize(sentence)

    for token in tokens:
        freq_dict[token] += 1

filtered_freq = {
    word: count for word, count in freq_dict.items()
    if word not in ["this", "with", "from", "that"]
}
top_words = sorted(
    filtered_freq.items(),
    key=lambda x: (x[1], len(x[0])),
    reverse=True
)[:10]
rare_words = sorted(
    filtered_freq.items(),
    key=lambda x: (x[1], -len(x[0]))
)[:5]
print("\n--- Word Importance Report ---\n")

print("Top Impact Words:")
for word, count in top_words:
    print(f"{word} → {count}")

print("\nRare but Informative Words:")
for word, count in rare_words:
    print(f"{word} → {count}")
```

**Task 6: Build Full Pipeline**

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, cleaned = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(cleaned)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
result = full_pipeline(sentences)
print("\n--- FINAL PIPELINE OUTPUT ---\n")
print("All Tokens:")
print(result["tokens"])

print("\nClean Sentences:")
for i, sent in enumerate(result["clean_sentences"], 1):
    print(f"{i}. {sent}")
```

**Task 7: Error Handling**

```python
def preprocess_text(text):
    if not text or not isinstance(text, str):
        return [], ""
    text = text.lower()

    import re
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'\S+@\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'(.)\1+', r'\1', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    words = text.split()
    words = [w for w in words if len(w) > 2 or w in ["no", "not"]]

    cleaned = " ".join(words)
    tokens = cleaned.split() if cleaned else []

    return tokens, cleaned
def test_error_handling():

    test_cases = {
        "Empty String": "",
        "Only Emojis": "😍😍😍😍",
        "Only Numbers": "1234567890",
        "Spaces Only": "     ",
        "None Value": None
    }

    print("\n--- Task 7: Error Handling Results ---\n")

    all_passed = True

    for name, input_text in test_cases.items():
        tokens, cleaned = preprocess_text(input_text)
        if tokens == [] and cleaned == "":
            status = "Handled"
        else:
            status = "Not Handled"
            all_passed = False
        print(f"Test Case: {name}")
        print(f"  Input:  '{input_text}'")
        print(f"  Output: '{cleaned}'")
        print(f"  Status: {status}")
        print("-" * 40)
    if all_passed:
        print("\n All Edge Cases Passed! The Preprocessing Engine is robust.")
    else:
        print("\n Some test cases failed. Please check the pipeline.")

test_error_handling()
```
